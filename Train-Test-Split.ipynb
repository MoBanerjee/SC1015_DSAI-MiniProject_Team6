```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

### Importing the original datasets

```python
maindataset = pd.read_csv('Final Datasets/dataCleaned.csv')
```

```python
maindataset.head()
```

```python
X = maindataset.drop(['popularity'], axis=1)
Y = maindataset['popularity']
```

```python
X
```

```python
Y
```

### Splitting the Original Dataset into Train and Test Data in *80:20* ratio. 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X_train['popularity'] = Y_train
X_test['popularity'] = Y_test
```

```python
X_train.info()
```

```python
X_test.info()
```

### Creating **csv** files for the train and test datsets.

```python
X_train.to_csv('Test Train Data/Train.csv')
X_test.to_csv('Test Train Data/Test.csv')
```
